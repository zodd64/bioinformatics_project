```python
labels = data.index
```

```python
labels
```

```python
for x in range(len(labels)):
    if(labels[x].startswith("ALL")):
        labels.values[x] = "ALL"
    elif(labels[x].startswith("AML")):
        labels.values[x] = "AML"
labels
```

### Split 75-25

```python
split = 0.75
```

```python
msk = np.random.rand(72) < split
```

```python
trainingSet = data.loc[msk, :]
```

```python
trainingSet
```

```python
testSet = data.loc[~msk, :]
```

```python
testSet
```

```python
trainingClass = labels[msk]
testClass = labels[~msk]
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
```

```python
neigh.fit(trainingSet, trainingClass)
```

```python
pr = neigh.predict(testSet)
```

```python
for x in range(len(testClass)):
        print("Index of test sample ",x , " predicted as:", pr[x] , " real class is:" ,testClass[x])
```

```python
print("KNN accuracy: ",metrics.accuracy_score(testClass, pr))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(trainingSet, trainingClass)
```

```python
pr2 = clf.predict(testSet)
```

```python
for x in range(len(testClass)):
        print("Index of test sample ",x , " predicted as:", pr2[x] , " real class is:" ,testClass[x])
```

```python
print("Decision Tree accuracy: ",metrics.accuracy_score(testClass, pr2))
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
pr3 = gnb.fit(trainingSet, trainingClass).predict(testSet)
```

```python
for x in range(len(testClass)):
        print("Index of test sample ",x , " predicted as:", pr3[x] , " real class is:" ,testClass[x])
```

```python
print("Naive Bayes accuracy: ",metrics.accuracy_score(testClass, pr3))
```
